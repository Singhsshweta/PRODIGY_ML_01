```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
train_data_path = '/kaggle/input/house-prices-advanced-regression-techniques/train.csv'
train_data = pd.read_csv(train_data_path)
```

```python
selected_train_data = train_data[['SalePrice', 'GrLivArea', 'BedroomAbvGr', 'FullBath']]
```

```python
selected_train_data = selected_train_data.dropna()
```

```python
X_train = selected_train_data[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
y_train = selected_train_data['SalePrice']
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
test_data_path = '/kaggle/input/house-prices-advanced-regression-techniques/test.csv'
test_data = pd.read_csv(test_data_path)
```

```python
selected_test_data = test_data[['Id', 'GrLivArea', 'BedroomAbvGr', 'FullBath']]
```

```python
selected_test_data = selected_test_data.fillna(X_train.mean())
```

```python
X_test = selected_test_data[['GrLivArea', 'BedroomAbvGr', 'FullBath']]
```

```python
predictions = model.predict(X_test)
```

```python
submission = pd.DataFrame({'Id': selected_test_data['Id'], 'SalePrice': predictions})
```

```python
print(submission.to_markdown(index=False))
```

```python
# Save the submission to a CSV file in the working directory
submission.to_csv('/kaggle/working/submission.csv', index=False)
```
